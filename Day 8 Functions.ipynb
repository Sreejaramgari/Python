```python
#password strength checker
password = input("Enter your password: ")
if len(password) < 8:
    print("Password must be at least 8 characters.")
else:
    isupper = False
    islower = False
    isdigit = False
    isspecial = False
    for i in password:
        if i.isupper():
            isupper = True
        elif i.islower():
            islower = True
        elif i.isdigit():
            isdigit = True
        else:
            isspecial = True
    if isupper and islower and isdigit and isspecial is True:
        print("Strong Password")
    else:
        print("Weak Password")
```

```python
#sum of elements
def sum_of_elements(lst):
    sum = 0
    for i in lst:
        sum += i
    return sum
lst = [1, 2, 3, 4, 5]
res = sum_of_elements(lst)
print(res)
```

```python
#sum of elements using range
def sum_of_elements(lst):
    sum = 0
    for i in range(len(lst)):
        sum += lst[i]
    return sum
lst = [1, 2, 3, 4, 5]
res = sum_of_elements(lst)
print(res)
```

```python
def add(x, y):
    return x + y
def sub(x, y):
    return x - y
def multiply(x, y):
    return x * y
def division(x, y):
    if y == 0:
        return "error"
    else:
        return x / y
def calculator():
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    choice = int(input("Enter your choice:"))
    a = int(input("Enter a number: "))
    b = int(input("Enter a number: "))
    if choice == 1:
        result = add(a, b)
        print("Result:" , result)
        
    elif choice == 2:
        result = sub(a, b)
        print("Result:" , result)
        
    elif choice == 3:
        result = muliply(a, b)
        print("Result:" , result)
        
    elif choice == 4:
        result = division(a, b)
        print("Result:", result)
calculator()
```
