```python
#Types of arguments
1.positional Arguments
2.keyword arguments
3.Default Arguments
4.variable length of positional arguments - single *
5.variable length keyword arguments - Double **
```

```python
#Positional arguments
#substraction of two numbers
def sub(x, y):
    return x - y
b1 = int(input())
b2 = int(input())
print(sub(b1,b2))
print(sub(b2,b1))
```

```python
#Keyword Arguments
#substraction of two numbers
def sub(x, y):
    return x - y
b1 = int(input())
b2 = int(input())
print(sub(x = b1,y = b2))
print(sub(y = b2,x = b1))
```

```python
#sum of two numbers
def sum_of_two_elements(x, y):
    sum = x + y
    return sum
res = sum_of_two_elements(1, 20)
print(res)
```

```python
#as number of arguments are increasing ,so function defination also changes,so we move to default arguments or variabble length positional arguments
```

```python
#default arguments
def add(w,x,y = 0,z = 0):
    return w + x + y + z
b1 = int(input())
b2 = int(input())
b3 = int(input())
b4 = int(input())
print(add(b1, b2))
print(add(b1, b2, b3))
print(add(b1, b2, b3, b4))
```

```python
#multiply of two numbers
def multiply(a, b = 0):
    return a*b
x1 = int(input())
x2 = int(input())
print(multiply(x1,x2))
```

```python
#multiply of three numbers
def multiply(a, b = 0,c = 0):
    return a*b*c
x1 = int(input())
x2 = int(input())
x3 = int(input())
print(multiply(x1,x2,x3))

```

```python
#variable length positional arguments
#sum of four numbers
def add(x, *args):
    print(type(args))
    print(args)
    sum = 0
    for i in args:
        sum += i
    return sum

print(add(10, 20, 30, 40, 50))
```

```python
#variable length positional arguments
#always positional,keyword,variable length
def add(x, y = 0, *args):
    print(type(args))
    print(args)
    s = 0
    for num in args:
        s += num
    return s

print(add(10, 20, 30, 40, 50))
```

```python
#variable length positional arguments
#always positional,keyword,variable length(default parameters)
def add(x, y = 0,z = 0, *args):
    print(type(args))
    print(args)
    s = 0
    for num in args:
        s += num
    return s

print(add(10, 20, 30, 40, 50))
```

```python
#Variable length Keyword Arguments
def add(x, y = 0, *args, **kwargs):
    print(x, y)
    print(type(args), args)
    print(type(kwargs), kwargs)
print(add(10,20,30,40,50,1,2,3,4,5, a = 10, b = 20, c = 30))
print(add(10))
print(add(10, -20))
print(add(10,20,30,40,50))
```
