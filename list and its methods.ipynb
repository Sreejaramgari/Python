```python
#list methods
#append(ele):adds the element at end of the list
#extend(list):adds all elements in list at end
#insert(pos, ele):inserts the element at specified position
l1 = [1, 2, 3]
l1.append(4)
print(l1)
l1.extend([5, 6, 7])
print(l1)
l1.insert(3, 20)
print(l1)
l1.insert(100, 1000)
print(l1)
l1.insert(-2, 90)
print(l1)
```

```python
#removing elements from list
#remove(val):it removes the first occurance value from the list,otherwise value error
#pop():it removes and return end value from list,if list is empty it raises index error
#pop(position):it remove and return specified  position element from list
#clear():it removes all elements from the list
l1 = [1, 2, 3, 4, 5]
l1.remove(3)
print(l1)
print(l1)
l1.pop(3)
val = l1.pop()
print(val)
print(l1)
#l1.pop(5)#pop index out of range
#print(l1)
l1.clear()
print(l1)
```

```python
#index(ele):it returns the index of first occurance element,if element not in list it raises val error
l = [1, 2, 3, 4, 5, 4, 4]
val = l.index(4)
print(val)
#val = l.index(200)#value error
#print(val)
val = l.index(4, -1)
print(val)
val = l.index(4, 5)
print(val)
```

```python
#count(ele):it returns the total occurance of an element in a list
l = [1,2,3,4,2,2]
print(l.count(2))
```

```python
#built in functions of list
#min()
#max()
#sum()
#len()
l = [1,2,3,4,5,6]
print(max(l))
print(min(l))
print(sum(l))
print(len(l))
```
