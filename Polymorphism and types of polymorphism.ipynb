```python
#method overriding/runtime polymorphism
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def show(self):
        return f" My name is {self.name} and i am a {self.age} years old"
class employee(person):
    def __init__(self, empid, name, age):
        person.__init__(self,name ,age)
        self.empid = empid
    def show(self):
        return f" My name is {self.name} and my employee id is {self.empid}"
e = employee(1001, "Sreeja" , 21)
print(e.show())
    
```

```python
#Method Overloading(compile time polymorphism)
class math():
    def add(self, x, y):
        return x + y
    def add(self, x, y, z):
        return x + y + z
m = math()
print(m.add(5, 6))
print(m.add(7, 8, 9))
```

```python
#method overloading achieved through default arguments
class math():
    def add(self, a, b, c = 0, d = 0):
        return a + b + c + d
m = math()
print(m.add(5, 6))
print(m.add(7, 8, 9))
print(m.add(1,2,3,4))
```

```python
# 3.operator overloading
class Number:
    def __init__(self, value):
        self.value = value
    def __add__(self, other):# Overloading + operator
        return Number(self.value + other.value)
n1 = Number(10)#objects
n2 = Number(20)
n3 = n1 + n2
print(n3.value)
```

```python
# 4.Duck typing
class Duck:
    def speak(self):
        return "Quack"
class Dog:
    def speak(self):
        return "Bark"
def make_sound(animal):
    print(animal.speak())
d = Duck()
dog = Dog()
make_sound(d)    
make_sound(dog)  
```
