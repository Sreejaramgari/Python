```python
import re

def is_valid_date(date):
    pattern_31 = r'^(?:31([\/-])(0?[13578]|1[02])\1(?:(?:19|20)\d\d))$'
    pattern_30 = r'^(?:(?:29|30)([\/-])(0?[469]|11)\1(?:(?:19|20)\d\d))$'
    pattern_feb28 = r'^(?:(?:0?[1-9]|1\d|2[0-8])([\/-])0?2\1(?:(?:19|20)\d\d))$'
    pattern_feb29 = r'^(?:29([\/-])0?2\1(?:(?:19|20)(?:0[48]|[2468][048]|[13579][26])|2000))$'

    if re.match(pattern_31, date):
        return True
    elif re.match(pattern_30, date):
        return True
    elif re.match(pattern_feb29, date):
        return True
    elif re.match(pattern_feb28, date):
        return True
    else:
        return False

# Examples
print(is_valid_date("31/01/2024"))  # True
print(is_valid_date("30/04/2024"))  # True
print(is_valid_date("29/02/2024"))  # True (leap year)
print(is_valid_date("29/02/2023"))  # False
```

```python
import re
def is_valid_date(date):
    #  (DD-MM-YYYY)
    pattern = r'^((0[1-9]|[12][0-9]|3[01])-(0[1-9]|1[0-2])-(\d{4}))$'
    match = re.fullmatch(pattern, date)
    if not match:
        return False
    day, month, year = map(int, match.groups()[1:])
    # Months with 31 days
    
    if month in [1, 3, 5, 7, 8, 10, 12]:
        return day <= 31
    # Months with 30 days
    
    elif month in [4, 6, 9, 11]:
        return day <= 30
    # February
    elif month == 2:
        # Leap year check
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return day <= 29
        else:
            return day <= 28

    return False
```

```python
#clock implementation
from IPython.display import clear_output
seconds_counter = minutes_counter = hours_counter = 0
while True:
    print(f"{hours_counter} : {minutes_counter} : {seconds_counter}") #f -strings inside numbers
    time.sleep(1)
    seconds_counter += 1
    if seconds_counter == 60:
        minutes_counter += 1
        seconds_counter = 0
        if minutes_counter == 60:
            hours_counter += 1
            minutes_counter = 0
            if hours_counter == 24:
                hours_counter = 0
    clear_output(wait = True)
                
```
