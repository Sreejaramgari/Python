```python
#Types of methods
#1.Instance Method
#2.Class Method
#3.Static
class Student:
    school_name = "Geetha High School"#class attribute ----same for Each object
    def __init__(self, name, age): # ----------------------> Initialization Constructor
        self.name = name #instance attribute -----different for each object
        self.age = age   #instance attribute
    def details(self):   # Instance method
        return f"My name is {self.name} and I am a {self.age} years old"

    @classmethod 
    def change_school_name(cls, new_name): #class method
        cls.school_name = new_name

    @staticmethod 
    def percentage(m1, m2, m3):#static method
        return (m1 + m2 + m3)/3
#Object Creation
s1 = Student("Sreeja", 21)
s2 = Student("Sowmya", 24)
print(s1.age)
s1.change_school_name("siddharta school")
print(s1.school_name)
print(s2.school_name)
Student.change_school_name("XYZ SCHOOL")
print(s1.school_name)
print(s2.school_name)
print(s1.percentage(85,86,87))
print(Student.percentage(85,86,87))
```

```python
#Encapsulation
#create bank access
class Bank:
    def __init__(self, name, email, amount = 1000):
        self.name = name
        self._email = email
        self.__amount = amount
        print("Constructer Executed")

    #string method
    def __str__(self):
        return f"The user name is {self.name} and email address is {self._email}"
    def withdraw(self, withdraw_amount):
        if self.__amount >= withdraw_amount:
            self.__amount -= withdraw_amount
            return "Withdrawal successful and current balance is {self.__amount}"
        return "Insufficient Balance"
    def __Total_Users(self):
        return "Total users in Bank is 102"
account1 = Bank("Sreeja","sreeja@gamil.com", 10000)
print(account1)
print(account1.name)
print(account1.withdraw(1000))
print(account1._email) #protected member access in outside the class
print(account1._Bank__Total_Users()) #private member access by using "Name Mangling" -accessing private member outside class"
```
