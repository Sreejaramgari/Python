```python
####set operations or set methods
#1. union(new_set):it returns all unique elements from sets
#2. intersection: it returns common elements
#3.difference (s1 - s2):it returns all the elements of s1 which are not in s2
#4.Symmetric difference: it returns non common elements
#subset,superset,disjoint are checking function returns true or false
s1 = {1,2,3,6}
s2 = {3,4,5,6}
print(s1.union(s2))
print(s1 | s2)
print("Ïntersection")
print(s1.intersection(s2))
print(s1 & s2)
print("Difference")
print(s1.difference(s2))
print(s1 - s2)
print("Symmetric Difference***********")
print(s1.symmetric_difference(s2))
print("Subset******")
print(s1.issubset(s2))
print({1,2}.issubset(s1))
print({1,2,5}.issubset(s1))
print("Superset********")
print(s2.issuperset(s1))
print(s1.issuperset({1,2}))
print({1,2,5}.issubset(s1))
print(set().issubset(s1))#every empty set is a subset
print("Disjoint***")
print(s1.isdisjoint(s2))
print(s1.isdisjoint({10}))
```

```python
#add(ele):it add to set,if the element is not present in set,otherwise nothing
#update(new_set):set elements updates with new set elements
#remove(ele):It removes the set element,if element not present in set it raises keyerror
#discard(ele):it removes the ele from set otherwise nothing it raises none
s = {1,2,3,4}
s.add(10)
print(s)
s.update({1,2,5,6})
print(s)
s.remove(10)
print(s)
s.remove(20)#**********key error
print(s)
s.discard(10)
print(s)
s.discard(5)
print(s)
```
