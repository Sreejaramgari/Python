```python
#if it does not contain any Exception or Without Exception
try:
    lst = [1,2,3,4]
    print(lst)
except:
    print('Something wrong in program')
else:
    print('Program Done')
finally:
    print('Successfully program Executed')
```

```python
#if it contain an error or With Exception
try:
    lst = [1,2,3,4]
    print(lst)
    print(lst[10])
except:
    print('Something wrong in program')
else:
    print('Program Done')
finally:
    print('Successfully program Executed')
```

```python
#Without Wxception - which block going to execute
try:
    a = 10
    b = 5
    print(a/b)
except:
    print("Something wrong in program")
else:
    print("Program done")
finally:
    print("Successfully program executed")
    
```

```python
#With Exception- which block going to execute
try:
    a = 10
    b = 0
    print(a/b)
except:
    print("Something wrong in program")
else:
    print("Program done")
finally:
    print("Successfully program executed")
    
```

```python
Case1:Corresponding except block will execute
try:
    a = 10
    b = 0
    print(a/b)
except ZeroDivisionError:
    print("Zero Division Not Possible")
except:
    print("Something wrong in program")
else:
    print("Program done")
finally:
    print("Successfully program executed")  
```

```python
#case2:Default except block will execute
try:
    a = 10
    b = 0
    a.append(10)
    print(a/b)
except ZeroDivisionError:
    print("Zero Division Not Possible")
except:
    print("Something wrong in program")
else:
    print("Program done")
finally:
    print("Successfully program executed")  
```

```python
#case3:
try:
    a = 10
    b = 0
    a.append(10)
    print(a/b)
except:
    print("Something wrong in program")
except ZeroDivisionError:
    print("Zero Division Not Possible")
else:
    print("Program done")
finally:
    print("Successfully program executed")  
```

```python
#case 4:
try:
    a = 10
    b = 0
    a.append(10)
    print(a/b)
except ZeroDivisionError:
    print("Zero Division Not Possible")
else:
    print("Program done")
finally:
    print("Successfully program executed")  
```

```python
#case 5:
try:
    a = 10
    b = 0
    a.append(10)
    print(a/b)
except AttributeError:
    print("Something wrong in program")
except ZeroDivisionError:
    print("Zero Division Not Possible")
else:
    print("Program done")
finally:
    print("Successfully program executed")  
```

```python
#case 7: Using Exception as a Class and print corresponding msg
try:
    a = 10
    b = 0
    print(a/b)
except Exception as e:
    print("Something wrong in program:", e)
```

```python
try:
    file = open("test1.txt", 'r')
    s = file.read()
    print(s)
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
```

```python
try:
    file = open("test1.txt", 'w')
    s = file.read()
    print(s)
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
```
