#String Operations
1.String Concatenation(using +)
2.String Repetation(using *)

```python
#Concatenation
s1 = 'Code'
s2 = 'gnan'
print(s1 + s2)
```

```python
#Concatenation
s1 = 'Code'
s2 = 'gnan'
print(s1 +" "+ s2)(#to get space)
print(s1 + s1)
print(s1 * 5)
print((s1 + " ") * 5)(#string repetition)
```

```python
#Attendance Management
n = 5
absent = 0
present = 0
for i in range(1, n+1):
    val = int(input(f"Enter Roll no {i} attendance (0 = absent, 1 = present): "))
    if val == 1:
        present += 1
    else: 
        absent += 1
        percentage = (present / n) * 100
print("Total Students:", n)
print("Total Present:", present)
print("Total Absent:", absent)
print("Attendance Percentage:", percentage, "%")
```

```python
#String Functions
#case conversiion functions
s = "Sreeja"
print(s.upper())
print(s.lower())
print(s.swapcase())
print(s.title())
print(s.capitalize())
```

```python
#String manipulation functions
#replace():it replace the old string with new string
s = " Sreeja   "
print(s.replace("Sreeja", "Ramgari"))
#split():
print(s.split("j"))
#strip()
print(s.strip())
#join()
lst = ['my','name','is','sreeja']
''.join(lst)
```

```python
#replace()
s = "The sky is blue and ocean is blue"
res = s.replace("blue", "red")
print(res)
```

```python
#split()
s = "The sky is blue and ocean is blue"
res = s.split()
print(res)
print(s.split(sep = "blue"))
print(s.split(sep = "Blue"))
```

```python
#index(sub_str):it returns the first occurence of  substring index if it is present in string othervise value error
s = "The sky is blue and ocean is blue"
print(s.index('is'))
print(s.index('z'))
```

```python
#find(sub_str):it returns the first occurence of  substring index if it is present in string othervise -1
s = "The sky is blue and ocean is blue"
print(s.find('is'))
print(s.find('z'))
```

```python
#Strip():it removes the leading and lagging white space
s = "        The sky is blue and ocean is blue           "
print(s.strip())
print(s.lstrip())
print(s.rstrip())


```

```python
#join
#'sep'.join(iterable): it joins the all elements in the iterable with separator
lst = ['The', 'sky', 'is', 'blue', 'and', 'ocean', 'is', 'blue']
print(''.join(lst))
print(' '.join(lst))
print('@'.join(lst))



```

```python
#String Checking functions
s = "sreeja3"
print(s.isupper())
print(s.islower())
print(s.isdigit())
print(s.isalpha())
print(s.isalnum())
```
