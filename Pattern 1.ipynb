```python
#* hallow right angle triangle
#* *
#*   *
#* * * *
n = 4
for i in range(n):
    for j in range(n):
        if j == 0 or i == j or i == n - 1:
            print("*" , end = " ")
        else:
            print(" " , end = " ")
    print()   
```
