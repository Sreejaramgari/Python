```python
#TYPES OF ATTRIBUTES
#1.CLASS ATTRIBUTES
#Class Definition
class Student:
    school_name = "Geetha High School"#class attribute ----same for Each object
    def __init__(self, name, age): # ----------------------> Initialization Constructor
        self.name = name #instance attribute -----different for each object
        self.age = age   #instance attribute
    def details(self):   #method
        return f"My name is {self.name} and I am a {self.age} years old"
#Object Creation
s1 = Student("Sreeja", 21)
s2 = Student("Sowmya", 24)
print(s1.name)
print(s2.name)
print(s1.details())
print(s1.school_name)
print(s2.school_name)
```

```python
#Calculator
class Calculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def details(self):
        return f"Addition{self.x + self.y}"
        return f"Substraction{self.x - self.y}"
s1 = Calculator(10, 2)
s2 = Calculator(8, 2)
print(s1.x)
print(s2.y)


```
