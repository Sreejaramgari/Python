```python
#Abstraction
from abc import ABC, abstractmethod
class shapes(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass
class cube(shapes):
    def __init__(self, a):
        self.a = a
    def area(self):
        return  6*self.a**2
    def perimeter(self):
        return 12*self.a
    def volume(self):
        return self.a**3
    def faces(self):
        return 6
c = cube(3)
print("area:", c.area())
print("Perimeter:", c.perimeter())
print("Volume:", c.volume())
print("faces:", c.faces())

        
```

```python
from abc import ABC, abstractmethod
class Shapes(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass

class square(Shapes):
    def __init__(self,side):
        self.side=side
    def area(self):
        return (self.side)**2
    def perimeter(self):
        return 4 * self.side


class Rectangle(Shapes):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        return self.length*self.breadth
    def perimeter(self):
        return 2 * (self.length + self.breadth)
class Triangle(Shapes):
    def __init__(self, a, b, c, height, base):
        self.a = a
        self.b = b
        self.c = c
        self.height = height
        self.base = base
    def area(self):
        return 0.5 * self.base * self.height

    def perimeter(self):
        return self.a + self.b + self.c
class Circle(Shapes):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius * self.radius
    def perimeter(self):
        return 2 * 3.14 * self.radius
s=square(2)
print(s.area())
print(s.perimeter())

r=Rectangle(3, 4)
print(r.area())
print(r.perimeter())

t=Triangle(2,3,4,5,6)
print(t.area())
print(t.perimeter())

c=Circle(4)
print(c.area())
print(c.perimeter())
```

```python
#abstarction
from abc import ABC, abstractmethod
class Shapes(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Cylinder(Shapes):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    def area(self):
        return 2 * 3.14 * self.radius * (self.radius + self.height)
    def perimeter(self):
        return 2 * 3.14 * self.radius   
    def volume(self):
        return 3.14 * self.radius**2 * self.height
        
class Sphere(Shapes):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 4 * 3.14 * self.radius**2
    def perimeter(self):
        return 2 * 3.14 * self.radius   
    def volume(self):
        return (4/3) * 3.14 * self.radius**3

cy = Cylinder(3, 5)
print(cy.area())
print(cy.perimeter())
print( cy.volume())

s = Sphere(3)
print(s.area())
print( s.perimeter())
print( s.volume())
```

```python
class Employee():
    def __init__(self,emp_id,name):
        self.emp_id = emp_id
        self.name = name
    def calculate_salary(self):
        pass
class SalariedEmployee(Employee):
    def __init__(self,emp_id,name,monthly_salary):
        super().__init__(name,emp_id)
        self.monthly_salary = monthly_salary
    def calculate_salary(self):
        return self.monthly_salary
class HourlyEmployee(Employee):
    def __init__(self,emp_id,name,hourly_rate,hours_worked):
        super().__init__(self,emp_id,name)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked
    def calculate_salary(self):
        return self.hours_rate * self.hours_worked
class CommissionEmployee(Employee):
    def __init__(self,emp_id,name,base_salary,commission_rate,total_sales):
        super().__init__(name,emp_id)
        self.base_salary = base_salary
        self.commission_rate = commission_rate
        self.total_sales = total_sales
    def calculate_salary(self):
        return self.base_salary + self.commission_rate * self.total_sales
E1 = SalariedEmployee(10001,"Sreeja",20000,500)
#E2 = HourlyEmployee(2,"Shravya",200, 120)
#E3 = CommissionEmployee(3, "Reena",15000,0.1,50000)
print(E1.calculate_salary())
#print(E2.calculate_salary())
#print(E3.calculate_salary())

    
```
