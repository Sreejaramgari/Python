```python
#DICTIONARIES
#dictionary is a ordered collection
details = {'name' : "Sreeja","Batch" : 49,"course" : "python"}
print(details)
```

```python
#DICTIONARIES
#dictionary is a ordered collection if we add duplicate keys again then first key value is replaced with new value 
#dictinaries are unique
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari"}
print(details)
```

```python
#DICTIONARIES
#dictionary is a ordered collection if we add duplicate keys again then first key value is replaced with new value 
#dictinary values can access using keys
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari"}
print(details['name'])
```

```python
#dictionaries does not support indexing
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari"}
print(details[0])
```

```python
#dictionary does not supports indexing and slicing
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari",0:1}
print(details[0])
```

```python
#update dictionary value and items
#dictionary is mutable
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari"}
details['course'] = 'java'
details['age'] = 22
print(details)
```

```python
#dictionary methods
#get(key, default_value): it returns the values if the key is present in the dictionary 
#            otherwise it returns default value
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari"}
details['location']
```

```python
#dictionary methods
#get(key, default_value): it returns the values if the key is present in the dictionary 
#            otherwise it returns default value
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari"}
print(details.get('location','india'))
print(details.get('name',"ram"))
print(details.get("marks"))
```

```python
#update(items):dictionary updates with new items
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari",0:1}
new_dict = {'course':'java','Batch':50,'location':'india'}
details.update(new_dict)
print(details)
```

```python
#keys():it returns the list of dictionary keys
#values():it returns the list of dictionary values
#items():it returns the dictionary list of tuple of key-value pairs
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari",0:1}
print(details.keys())
print(details.values())
print(details.items())
print(list(details.items()))
```

```python
#keys():it returns the list of dictionary keys
#values():it returns the list of dictionary values
#items():it returns the dictionary list of tuple of key-value pairs
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari",0:1}
print(type(details.keys()))
print(type(details.values()))
print(type(details.items()))
print(list(details.items()))
```

```python
#pop(key):it returns the values and removes the item from dictionary if key is present ,otherwise keyerror
#popitem():it returns and removes last item
details = {'name' : "Sreeja","Batch" : 49,"course" : "python", "name": "Ramgari",0:1}
print(details.pop('name'))
print(details.popitem())
print(details)
```
